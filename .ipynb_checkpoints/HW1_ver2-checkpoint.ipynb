```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
import seaborn as sns
from colour import Color
from json import JSONEncoder
```

```python
time_df = pd.read_csv('./Rawdata/DailyTemperatureDaejeon.csv')
space_df =gpd.read_file('./Rawdata/DailyTemperatureStation.csv')
crs = {'init': 'epsg:4326'}

print(time_df.head())
print(space_df.head())

type(time_df)
```

```python
'''
Visualize climate data by time seriese

Using simple plotting, 
visualize average daily temperature, maximum and minimum of each day at once.

x axis: date (daily)
y axis: temperature
'''

time_df.plot(x =  'Date', y = ['Maximum', 'Daily temperature', 'Minimum'], colormap = 'RdYlBu', figsize = (14, 10),marker = '+', markersize = 5)
xlab = 'Date'
ylab = 'Temperature'
title = 'Time series data'

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

plt.show()
```

```python
'''
Visualize categorical data on geographical map

Using simple plotting,
visualize the locatino of each data with categorical classification based on administrative district (do)

x axis: Longitude
y axis: Latitude
color: categorization for administrative district
'''

numeric_list = ['Latitude', 'Longitude', 'Altitude (m)']
for i in numeric_list:
    space_df[i] = space_df[i].astype(float)

geometry = [Point(xy) for xy in zip(space_df["Longitude"], space_df["Latitude"])]
new_space_df = gpd.GeoDataFrame(space_df, crs = crs, geometry = geometry)
new_space_df.head()

do_list = new_space_df['Location'].unique()
colormap = "RdYlBu_r"
fig, ax = plt.subplots(figsize = (15, 15))
ax.grid()
new_space_df.plot(column = 'Location',ax = ax, markersize = 500, cmap = colormap, marker = 'o', legend = True)
xlab_2 = 'Longitude'
ylab_2 = 'Latitude'
title_2 = 'Categorical data based on location'

plt.xlabel(xlab_2)
plt.ylabel(ylab_2)
plt.title(title_2)
plt.show()
```
